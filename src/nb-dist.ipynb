```julia
using Distributions
using Plots
```

```julia
# 確率pで成功する独立なベルヌーイ試行がr回成功するまでに失敗する回数xの確率
# 成功回数
r = 3
# 成功する確率
p = 0.5
# 失敗する回数
xs = 1:10
bar(xs, pdf(NegativeBinomial(r, p), xs), xlabel="x", ylabel="Dens")
```

```julia
# 負の二項分布に従うデータを生成
data = rand(NegativeBinomial(r, p), 100)
histogram(data; bins=20, xlabel="x", ylabel="Freq")
```
