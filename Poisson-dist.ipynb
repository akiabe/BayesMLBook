## 「ベイズ推論による機械学習入門」P.55 ポアソン分布 (Poisson distribution)

- 非負の整数x を生成するための確率分布
- 定義

$$Poi(x|\lambda) = \frac{\lambda^x}{x!}e^{-\lambda}$$


$$\lambda \in R^{+}$$

- 対数表示

$$ lnPoi(x|\lambda) = xln\lambda - lnx! - \lambda$$

- 期待値

$$\langle x \rangle = \lambda $$
$$\langle x^2 \rangle = \lambda(\lambda + 1) $$

### ポアソン分布

```julia
Poi(x, λ) = λ^x/factorial(x) * exp(-λ)
```

### パラメータの設定を変えた確率質量関数のグラフ

```julia
# x軸設定
xs = 0:1:10

# λ = 0.5
p1 = Poi.(xs, 0.5)

# λ = 4.0
p2 = Poi.(xs, 4.0)

# λ = 1.0
p3 = Poi.(xs, 1.0)

# 作図
using Plots
plot(bar(xs, p1, label="λ=0.5"),
    bar(xs, p2, label="λ=4.0"),
    bar(xs, p3, label="λ=1.0"))
```
