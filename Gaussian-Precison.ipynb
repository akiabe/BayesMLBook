```julia
# Train data
using Distributions
N = 50; μ = 25; λ = 0.01
Xᵗ = rand(Normal(μ,√(1/λ)), N)

# Hyperparameter
a = 1; b = 1

# Training
â, b̂ = (N/2+a, 0.5*sum((Xᵗ.-μ).^2)+b)

# Test data
Xᵖ = range(0, 40, length=50)

# Prediction
μₚ, λₚ, υₚ = (μ, â/b̂, 2*â)

# Plot
using SpecialFunctions, PyPlot
St(x, μ, λ, υ) = gamma((υ+1)/2)/gamma(υ/2)*(λ/(π*υ))^0.5*(1+λ/υ*(x-μ)^2)^(-(υ+1)/2)
bar(Xᵖ, St.(Xᵖ, μₚ, λₚ, υₚ));
```
