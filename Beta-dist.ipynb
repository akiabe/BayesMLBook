## 「ベイズ推論による機械学習入門」P.57 ベータ分布 (beta distribution)

- 定義
$$
Beta(\mu|a,b) = C_B(a,b)\mu^{a-1}(1-\mu)^{b-1}\\
$$

- パラメータ
$$
a \in R^+ \quad b \in R^+\\
$$

- 正規化項
$$
C_B(a,b) = \frac{\Gamma(a+b)}{\Gamma(a)\Gamma(b)}\\
$$

- 対数表示
$$
lnBeta(\mu|a,b) = (a-1)ln\mu + (b-1)ln(1-\mu) + lnC_B(a,b)\\
$$

- 期待値
$$
  \langle \mu \rangle = \frac{a}{a+b} \quad \langle ln\mu \rangle = \psi(a) - \psi(a+b) \\
$$

### ベータ分布

```julia
using SpecialFunctions: gamma

const Γ = gamma

Beta(μ,a,b) = Γ(a+b)/(Γ(a)*Γ(b)) * μ^(a-1)*(1-μ)^(b-1)
```

### パラメータの設定を変えた時の確率密度のグラフ

```julia
# μを設定
μ = 0.0:0.01:1.0

# a = 0.5, b = 0.5
p1 = Beta.(μ, 0.5, 0.5)

# a = 0.6, b = 0.8
p2 = Beta.(μ, 0.6, 0.8)

# a = 1.0, b = 1.0
p3 = Beta.(μ, 1.0, 1.0)

# a = 10.0, b = 40.0
p4 = Beta.(μ, 10.0, 40.0)

# a = 10.0, b = 5.0
p5 = Beta.(μ, 10.0, 5.0)

using Plots
plot(plot(μ, p1, label="a=0.5,b=0.5"), plot(μ, p2, label="a=0.6,b=0.8"), 
    plot(μ, p3, label="a=1.0,b=1.0"), plot(μ, p4, label="a=10.0,b=10.0"), 
    plot(μ, p5, label="a=10.0,b=5.0"))
```
