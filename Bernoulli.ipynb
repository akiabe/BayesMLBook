```julia
# Train data
using Distributions
N = 50
Xᵗ = rand(Bernoulli(0.25), N)

# Hyperparameter
a = 0.5
b = 0.5

# Training
â = sum(Xᵗ) + a
b̂ = N - sum(Xᵗ) + b

# Prediction
using SpecialFunctions
f(a, b, x) = exp(loggamma(a+b)+loggamma(x+a)+loggamma(1-x+b) -loggamma(a)-loggamma(b)-loggamma(a+b+1))

# plot
using PyPlot
bar([0, 1], [f(â, b̂, 0), f(â, b̂, 1)]);
```
