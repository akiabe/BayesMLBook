```julia
# Train data
using Distributions
N = 50; μ = 25; λ = 0.01
Xᵗ = rand(Normal(μ,√(1/λ)), N)

# Hyperparameter
m, β, a, b = (0, 1, 1, 1)

# Training
β̂ = N+β; m̂ = (sum(Xᵗ)+β*m)/β̂
â = N/2+a; b̂ = 0.5*(sum(Xᵗ.^2)+β*m^2-β̂*m̂^2)+b
 
# Test data
Xᵖ = range(-50, 50, length=50)

# Prediction
μₚ, λₚ, υₚ = (m̂, (β̂*â)/((1+β̂)*b̂), 2*â)

# Plot
using SpecialFunctions, PyPlot
St(x, μ, λ, υ) = gamma((υ+1)/2)/gamma(υ/2)*(λ/(π*υ))^0.5*(1+λ/υ*(x-μ)^2)^(-(υ+1)/2)
bar(Xᵖ, St.(Xᵖ, μₚ, λₚ, υₚ));
```
