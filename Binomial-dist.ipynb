```julia
### ベイズ推論による機械学習 2.2.2節二項分布の実装

## 二項分布 

# M:コイントスを行った回数, m:表が出た回数, μ:パラメータとする
Bin(m, M, μ) = (factorial(M)*inv(factorial(m)*factorial(M-m))) * μ^m*(1-μ)^(M-m)

# コイントスを10回行い、表が5回出る確率
println("コイントスを10回行い、表が5回出る確率: ", Bin(5, 10, 0.5))

## 様々なパラメータM及びμを設定した時の分布

# M=10, μ=0.5
p1 = zeros(10)
for i in 1:10
    p1[i] = Bin(i, 10, 0.5)
end

# M=10, μ=0.2
p2 = zeros(10)
for i in 1:10
    p2[i] = Bin(i, 10, 0.2)
end

# M=20, μ=0.6
p3 = zeros(20)
for i in 1:20
    p3[i] = Bin(i, 20, 0.6)
end

# 作図
using Plots
plot(bar(1:10, p1, label="M=10,μ=0.5"), 
    bar(1:10, p2, label="M=10,μ=0.2"), 
    bar(1:20, p3, label="M=20,μ=0.6"))
```
